## Nettoyage de nos données et quelques statistiques

Tout d'abord, on va compter le nombre de lignes qu'on a :

````python
print(f"Il y a {metadata_raw.shape[0]} séquences dont on a les métadonnées avant le nettoyage.")
````

On va maintenant changer les noms des colonnes. Il vaut toujours mieux éviter d'avoir des noms de colonnes avec des espaces.

````python
metadata_raw.rename(columns={'Accession': 'accession',
                   'Geographic Location': 'country_and_region',
                   'Geographic Region': 'continent',
                   'Completeness': 'completeness',
                   'Length': 'length',
                   'Virus Pangolin Classification': 'lineage',
                   'Isolate Collection date': 'date',
                  },
          inplace=True
         )

print("On a changé le nom des colonnes. Regardons ensemble :")
metadata_raw.head(10)
````

On va maintenant évaluer le nombre de valeurs manquantes qu'on compte dans notre BDD.

````python
# On compte le nombre de valeurs "NaN" par colonne
print("On compte le nombre de valeurs 'NaN' par colonne")
metadata_raw.isna().sum()
````

Aucune colonne ne compte de valeurs manquantes, notre BDD semble très complète.

On va maintenant compter le nombre de lignes pour lesquelles la date n'est pas dans le format 'yyyy-mm-dd' (ex : 2022-08-26).

````python
metadata_raw['date'] = metadata_raw.date.where(metadata_raw.date.str.match(pat='[0-9][0-9][0-9][0-9]-[0-9][0-9]-[0-9][0-9]'))
print("On compte le nombre de lignes pour lesquelles le format de date n'est pas yyyy-mm-dd'  :")
metadata_raw.isna().sum()
````

Il y a 15 lignes n'ayant pas le format de date de prélévement voulu, on va les supprimer.

````python
metadata_raw.dropna(inplace=True)
print(f"Il y a encore {metadata_raw.shape[0]} séquences dans le dataset après nettoyage.")
print(f"Les échantillons ont été prélevés entre {metadata_raw.date.sort_values().head(1).to_string(index=False)} et {metadata_raw.date.sort_values().tail(1).to_string(index=False)}.")
````

On va créer une colonne *month* et une colonne *year* à partir de la colonne *date*, cela pourra nous servir plus tard.

````python
metadata_raw["year"] = metadata_raw["date"].str.split(pat='-').str[0] #création de la colonne "year"
metadata_raw["month"] = metadata_raw["date"].str.split(pat='-').str[1] #création de la colonne "month"

metadata_raw['date'] = pd.to_datetime(metadata_raw['date'], format='%Y-%m-%d') #on convertit la colonne date en objet datetime
print("On a maintenant les colonnes year et month :")
metadata_raw.head(10)
````

Certaines valeurs de country_and_region se présentent sous le format suivant : « Pays:Région ». Nous les avons réparties dans deux colonnes différentes. Pour les colonnes où la région n'est pas définie, la valeur « None » a été attribuée.

````python
metadata_raw[['country', 'region']] = metadata_raw['country_and_region'].str.split(':', expand=True)
metadata_raw = metadata_raw.drop("country_and_region", axis='columns')
metadata_raw[metadata_raw['region'].notnull()].head(10)
````

Quelques statistiques maintenant !

On va maintenant afficher la première séquence détectée en Afrique pour chacune de nos lignées.

````python
lineages_first_detected = metadata_raw.groupby("lineage", #on groupe nos données par "lineage"
                                              as_index=False).apply(
    lambda df: df.sort_values("date").head(1)             #on ne prend que la séquence dont la date est la plus ancienne
                                              )

print(f"Voici la liste des premières séquences détectées en Afrique pour chacune de nos {lineages_first_detected.shape[0]} lignées.")
lineages_first_detected
````

On va maintenant afficher quelques informations par lignée :

````python
lineages_data = metadata_raw.groupby("lineage", #on groupe par lignée
    as_index = False).agg({                     #on réalise quelques opératios par lignée comme :
    'country' : lambda x : x.value_counts().idxmax(), #pays avec le plus de séquences
    'length' : "mean",                                #longueur moyenne des génomes séquencés en nucléotides
    'date' : lambda x : x.sort_values().head(1),      #date de découverte de la première séquence de la lignée
    'accession' : "count",                            #nombre de séquences 
    'year' : lambda x : x.value_counts().idxmax()     #année où l'on retrouve le plus cette séquence
}).set_axis(['lineage',                         # on renomme les colonnes
             'most_represented_country',
             'mean_sequences_length', 
             'first_detected', 
             'genomes_number',
             'most_represented_year'], 
            axis=1)

print("Voici quelques informations par lignées, triées par nombre décroissant de génomes séquencés :")
lineages_data.sort_values("genomes_number", ascending=False)
````

## Modélisation des données

Maintenant qu'on a vu qu'on pouvait obtenir quelques statistiques sur notre dataset, on va essayer de faire ressortir les éléments importants d'une manière plus lisible. On va commencer par faire quelques préparations avant de les modéliser. 

### Quelques préparations

#### Fonctions utiles

````python
# cette fonction prend en entrée un fichier et une liste
# elle écrit dans le fichier de sortie un élément de la liste par ligne
def write_list_to_file(output_file, input_list) :
    with open(output_file, 'w') as f:
        for element in input_list:
            f.write(f"{element}\n")
````

#### Assigner des couleurs aux lignées

Afin que l'on puisse faire des graphes et que les couleurs de nos lignées restent à chaque fois les même, on va créer des couleurs de façon aléatoire. Pour cela, on va utiliser la librairie *distinctipy*. On va faire de même pour générer des couleurs aux pays mentionnés par nos données.

````python
# nombre de couleurs à générer (autant que de lignées)
N = len(lineages_list)

# on génère N couleurs
distinctipy_lineages_colors = distinctipy.get_colors(N)

print("Voici les couleurs choisies pour nos lignées")
# on affiche les couleurs
distinctipy.color_swatch(distinctipy_lineages_colors)
````

````python
# nombre de couleurs à générer (autant que de pays)
N = len(countries_list)

# on génère N couleurs mais différentes des couleurs déjà choisies pour nos lignées + noir et blanc
distinctipy_countries_colors = distinctipy.get_colors(N, distinctipy_lineages_colors + [(0,0,0), (1,1,1)])

print("Voici les couleurs choisies pour nos pays")
# on affiche les couleurs
distinctipy.color_swatch(distinctipy_countries_colors)
````

On va transformer ces couleurs d'un format rgb à un format en hexadécimal.

````python
from matplotlib.colors import rgb2hex

hex_distinctipy_lineages_colors = [ rgb2hex(x) for x in distinctipy_lineages_colors ]

hex_distinctipy_countries_colors = [ rgb2hex(x) for x in distinctipy_countries_colors ]

print("Voici nos couleurs assignées aux lignées sous format hexadécimal :")
print(hex_distinctipy_lineages_colors)
````

Le problème, c'est qu'à cheque fois qu'on exécute les cellules précédentes, on génère de nouvelles couleurs aléatoires. On va donc charger celles qui ont déjà été définies. Si elles ne le sont pas encore, on va utiliser celles qu'on vient de produire.

````python
#get or save lineages domain (lineages)
fixed_lineages_domain_file = data_repo + "fixed_lineages_domain.txt"
fixed_domain_lineages = None

if not os.path.exists(fixed_lineages_domain_file):
    fixed_domain_lineages = list(metadata_raw["lineage"].unique())
    write_list_to_file(fixed_lineages_domain_file, fixed_domain_lineages)
else :
    with open(fixed_lineages_domain_file) as f :
        fixed_domain_lineages = f.read().splitlines()

#get or save lineages range (colors)
fixed_lineages_range_file = data_repo + "fixed_lineages_range.txt"
fixed_range_lineages = None

if not os.path.exists(fixed_lineages_range_file):
    fixed_range_lineages = hex_distinctipy_lineages_colors
    write_list_to_file(fixed_lineages_range_file, fixed_range_lineages)
else :
    with open(fixed_lineages_range_file) as f :
        fixed_range_lineages = f.read().splitlines()
````

````python
#get or save countries domain (countries)
fixed_countries_domain_file = data_repo + "fixed_countries_domain.txt"
fixed_domain_countries = None

if not os.path.exists(fixed_countries_domain_file):
    fixed_domain_countries = list(metadata_raw["country"].unique())
    write_list_to_file(fixed_countries_domain_file, fixed_domain_countries)
else :
    with open(fixed_countries_domain_file) as f :
        fixed_domain_countries = f.read().splitlines()

#get or save countries range (colors)
fixed_countries_range_file = data_repo + "fixed_countries_range.txt"
fixed_range_countries = None

if not os.path.exists(fixed_countries_range_file):
    fixed_range_countries = hex_distinctipy_countries_colors
    write_list_to_file(fixed_countries_range_file, fixed_range_countries)
else :
    with open(fixed_countries_range_file) as f :
        fixed_range_countries = f.read().splitlines()
````

#### A quoi ressemble notre carte ?

On va maintenant, grâce à la librairie *Altair*, afficher notre carte de l'Afrique et colorier par la couleur qu'on a choisit précédemment les pays pour lesquels on a des données. 

````python
input_data_total = metadata_raw
continent_map = africa_map.copy()
input_data_total = metadata_raw[["accession", "country"]]


continent_map = africa_map.copy()

# création de notre objet
basemap = (
    alt.Chart(continent_map) #on va travailler sur notre GeoDataFrame d'Afrique
    .mark_geoshape(
        stroke='white', strokeWidth=0.5 #on affiche les délimitations des pays par des lignes blanches
    ).transform_lookup( #on va "merge" notre GeoDataFrame à notre dataset...
        lookup="name", #... par le nom du pays...
        from_=alt.LookupData(input_data_total, "country", ["accession"]) #... en récupérant juste l'info sur son accession number
    ).encode(
        color=alt.condition('datum.accession !== null', #on colorie notre carte si le pays à des séquences
                            alt.Color( 
            'country:N', legend=None                     
       ).scale(
            domain=fixed_domain_countries, range=fixed_range_countries # en utilisant les couleurs définies plus tôt
        ),
                            alt.value('lightgray')) # si le pays n'a pas de séquence, il apparaît en gris

    )
)

#on affiche notre carte
basemap.display() 
````

### Faire un diagramme en barre des lignées prédominantes par période

On va une fois de plus utiliser la librairie Altair qui va nous permettre de créer notre graphe.Ici, nous allons essayer de faire un diagramme interactif, c'est à dire qui change en fonction de l'option choisie. Nous allons donc afficher un diagramme en barre montrant les 10 lignées les plus prédominantes par période de temps choisie, classées par ordre décroissant. En gros, on choisit une période de temps (une année ou toutes les années) et on compte, pour chacune lignée, combien de séquences en font partie.

````python
output_barchart_lineages = Output()

# on créé notre menu déroulant
# qui prend comme valeur "All" (toutes les années), ou une année particulière
year_options_barchart = ["All"] + list(metadata_raw.year.unique()) 
year_dropdown_barchart = Dropdown(
    description='Year:',
    options=year_options_barchart
)

display(output_barchart_lineages)

# on le rend interactif
@interact(year = year_dropdown_barchart)

# on créé une fonction qui affiche notre diagramme en barre à chaque changement d'année
def display_barchart(year) :

    #on va travailler sur notre dataset
    input_data = metadata_raw

    if year!=None :
    
        with output_barchart_lineages :
            #
            output_barchart_lineages.clear_output()

            # si la période choisie est différent de "All", on choisit une réduction de notre dataset 
            # avec juste les séquences prélevées l'année choisie
            if year != "All" :
                input_data = input_data.loc[input_data['year'] == year,].copy()

            # on groupe les séquences par lignées
            total_count_africa = input_data.groupby(
                ["lineage"],
                as_index = False
            ).size() # et on compte le nombre de séquences par lignées

            #on renomme juste la colonne en "sequences_number"
            total_count_africa.rename(columns={
                'size' : 'sequences_number'
            },
                inplace=True)

            # on choisit les 10 lignées ayant les plus de séquences
            top10_selection = total_count_africa.nlargest(10, columns='sequences_number')

            # on créé une colonne supplémentaire "Others" faisant la somme des séquences dans des lignées
            # ne faisant pas partie du top 10
            top10_selection.loc[len(top10_selection)] = ['Others', 
                                                           total_count_africa.loc[
                                                           ~total_count_africa.lineage.isin(top10_selection.lineage), 
                                                           'sequences_number'
                                                          ].sum()]

            # on créé notre graphe
            barchart = alt.Chart(
                top10_selection #on travaille avec notre objet nouvellement créé
            ).mark_bar( # on indique qu'on fait un diagramme en barre
                
            ).encode(
                    alt.X('lineage:N', # selon l'axe X, on affiche les différentes lignées
                         title="Lineage", 
                        sort=None),
                    alt.Y('sequences_number:Q', # selon l'axe Y, on affiche le nombre de séquences par lignées
                         title="Sequenced genomes number"),
                    alt.Color(
                        "lineage:N" #on colorie nos barres en fonction de la lignée
                    ).scale(
                        domain=fixed_domain_lineages + ["Others"], 
                        range=fixed_range_lineages + ["#000000"] # en utilisant le code couleur définit précédemment
                    ).legend(title='Lineage',
                                 # notre légende affiche les noms des lignées de notre top 10 + "Others"
                                 values=list(top10_selection.lineage.unique())) 
            )
            
            # on définit un titre pour notre graphe
            barchart_title = "Top 10 predominant lineages in African countries in "
            # en fonction de l'année
            if year == "All" :
                barchart_title += "2020-2022 period"
            else:
                barchart_title += year

            # puis on affiche notre diagramme
            barchart.properties(
                title = barchart_title
            ).display()
````

En cliquant sur le menu déroulant, on peut choisir la période de temps ciblée et afficher le nombre de séquences par lignée trouvée pendant cette période en Afrique.

De plus, on peut voir en haut à droite de notre diagramme 3 points encerclés. En cliquant dessus on peut enregistrer notre graphe en tant que SVG (très bonne qualité, recommandé pour les publications, ou en PNG.

### Faire un diagramme en barre des lignées prédominantes par pays

On va faire la même chose mais cette fois ci en prenant en compte un pays. On affichera le nombre de séquences par lignée ayant été collectées dans ce pays pendant la totalité de la période.

````python
output_barchart_lineages_country = Output()

# on créé notre menu déroulant
# qui prend comme valeur "All" (tous les pays), ou un pays particulier
country_options_barchart = ["All"] + list(metadata_raw.country.sort_values().unique()) 
country_dropdown_barchart = Dropdown(
    description='Country:',
    options=country_options_barchart
)

display(output_barchart_lineages_country)

# on le rend interactif
@interact(country = country_dropdown_barchart)

# on créé une fonction qui affiche notre diagramme en barre à chaque changement de pays
def display_barchart_country(country) :

    #on va travailler sur notre dataset
    input_data = metadata_raw

    if country!=None :
    
        with output_barchart_lineages_country :
            #
            output_barchart_lineages_country.clear_output()

            # si le pays choisi est différent de "All", on choisit une réduction de notre dataset 
            # avec juste les séquences prélevées dans le pays choisi
            if country != "All" :
                input_data = input_data.loc[input_data['country'] == country,].copy()

            # on groupe les séquences par lignées
            total_count_africa = input_data.groupby(
                ["lineage"],
                as_index = False
            ).size() # et on compte le nombre de séquences par lignées

            #on renomme juste la colonne en "sequences_number"
            total_count_africa.rename(columns={
                'size' : 'sequences_number'
            },
                inplace=True)

            # on choisit les 10 lignées ayant les plus de séquences
            top10_selection = total_count_africa.nlargest(10, columns='sequences_number')

            # on créé une colonne supplémentaire "Others" faisant la somme des séquences dans des lignées
            # ne faisant pas partie du top 10
            top10_selection.loc[len(top10_selection)] = ['Others', 
                                                           total_count_africa.loc[
                                                           ~total_count_africa.lineage.isin(top10_selection.lineage), 
                                                           'sequences_number'
                                                          ].sum()]

            # on créé notre graphe
            barchart_per_country = alt.Chart(
                top10_selection #on travaille avec notre objet nouvellement créé
            ).mark_bar( # on indique qu'on fait un diagramme en barre
                
            ).encode(
                    alt.X('lineage:N', # selon l'axe X, on affiche les différentes lignées
                         title="Lineage", 
                        sort=None),
                    alt.Y('sequences_number:Q', # selon l'axe Y, on affiche le nombre de séquences par lignées
                         title="Sequenced genomes number"),
                    alt.Color(
                        "lineage:N" #on colorie nos barres en fonction de la lignée
                    ).scale(
                        domain=fixed_domain_lineages + ["Others"], 
                        range=fixed_range_lineages + ["#000000"] # en utilisant le code couleur définit précédemment
                    ).legend(title='Lineage',
                                 # notre légende affiche les noms des lignées de notre top 10 + "Others"
                                 values=list(top10_selection.lineage.unique())) 
            )
            
            # on définit un titre pour notre graphe
            barchart_title_country = "Top 10 predominant lineages in African countries in "
            # en fonction du pays
            if country == "All" :
                barchart_title_country += "Africa"
            else:
                barchart_title_country += country

            # puis on affiche notre diagramme
            barchart_per_country.properties(
                title = barchart_title_country
            ).display()
````

Les options sont nombreuses et seules notre façon d'imaginer représenter nos données représente une limite à ce qu'on peut faire. On aurait pu aller plus loin et pouvoir, sur un seul graphe, choisir à la fois la période de temps et le pays pour lequel on veut afficher le compte de séquences par lignées.

### Carte de l'Afrique et lignées prédominantes

On va maintenant afficher une carte de l'Afrique et colorier cette fois-ci les pays en fonction de la lignée prédominante trouvée dans le pays. De plus, on va afficher un cercle au centre de chaque pays dont la taille sera proportionnelle au nombre de séquences toutes lignées confondues, dans le pays. Pour cela, on va encore utiliser la librairie Altair.

````python
# tout d'abord on se créé le dataframe qu'on va utiliser ici

# on groupe les séquences par pays
lineages_by_country = metadata_raw.groupby("country").agg({
    'lineage' : lambda x : x.value_counts().idxmax(), # on relève la lignée la plus représentée dans le pays
    'accession' : "count"                             # on compte le nombre de séquences total provenant du pays
}).set_axis(['most_represented_lineage',              # on renomme nos colonnes
            'sequences_number'], 
            axis=1).reset_index()

print("Voici à quoi ressemble le dataset qu'on va utiliser :")
lineages_by_country
````

````python
# pour notre carte, on va utiliser le GeoDataFrame d'Afrique
continent_map = africa_map.copy()

# on créé la base de notre carte
basemap = (
    alt.Chart(continent_map)
    .mark_geoshape(
        stroke='white', strokeWidth=0.5 #on marque les frontières par des lignes blanches
    ).transform_lookup( # on associe notre GeoDataFrame à celui contenant nos valeurs (voir #1#)
        lookup="name", # on les associe par le nom des pays (voir #2#)
        from_=alt.LookupData(
            lineages_by_country, #1#
            "country", #2#
            ["most_represented_lineage", "sequences_number"] # on sélectionne les valeurs qu'on veut garder
        )
    ).encode(
        color=alt.condition(
            'datum.sequences_number !== null',  # on ne colorie que les pays pour lesquels on a des séquences
            alt.Color(
                'most_represented_lineage:N' # on colorie par lignée la plus représentée
           ).scale(
                domain=fixed_domain_lineages, range=fixed_range_lineages # pays coloriés selon l'échelle définie précédemment
            ).legend(title='First detected lineage',
                     values=list(lineages_by_country.most_represented_lineage.unique())),
                                alt.value('lightgray') # les pays sans valeurs sont en gris
        ),
        tooltip=[ # on ajoute une animation au passage de la souris sur un pays
      {"type": "nominal", 'title': 'pays',"field": "name"}, # on ajoute ce qu'on veut montrer donc le nom du pays
      {"type": "nominal", 'title': 'most represented lineage',"field": "most_represented_lineage"}, # la lignée la plus représentée
      {"type": "quantitative", 'title': 'genomes sequenced', "field": "sequences_number"} # le nombre de séquences
    ]

    )
)

continent_map_centroids = africa_centroids.copy()
# pour faire les cercles représentant le nombre de séquences du pays, on doit récupérer les coordonnées des
# centroids pour chaque pays
continent_map_centroids["lon"] = continent_map_centroids.geometry.x #on créé une colonne pour la longitude
continent_map_centroids["lat"] = continent_map_centroids.geometry.y #on créé une colonne pour la latitude

# on créé notre deuxième chart qui va venir se superposer sur notre carte de base
# ici notre chart n'est composé que de cercles situés aux coordonnées des centroïds des pays
bubbles = (
    alt.Chart(
        continent_map_centroids
    ).transform_lookup( # on associe notre GeoDataFrame à celui contenant nos valeurs (voir #3#)
        lookup="name", #on les associe par le nom des pays (voir #4#)
        from_=alt.LookupData(
            lineages_by_country, #3#
            "country", #4#
            ["most_represented_lineage", "sequences_number"] # on sélectionne les valeurs qu'on veut garder
        )
    ).transform_filter(
        datum.sequences_number != None # on filtre pour ne garder que les pays avec des séquences
    ).mark_circle( # on fait des cercles...
        color="red", # ...de couleur rouge...
        opacity=0.5, # ...un peu transparent ...
        filled=True # ... mais remplis ...
    ).encode(
        longitude="lon:Q", # ...qu'on situe par la longitude...
        latitude="lat:Q", #...et la latitude
        size=alt.Size("sequences_number:Q", # les cercles ont une taille en fonction du nombre de séquences
                      scale=alt.Scale(type="log"), # on utilise une échelle logarithmique
                      legend=alt.Legend( # on ajoute une légende 
                        title="Sequences number",
                        orient='none',
                        legendX=800, legendY=0,
                        direction='vertical')
                     ),
        tooltip=[ # on affiche la même animation que si on passe la souris sur le pays
      {"type": "nominal", 'title': 'pays',"field": "name"},
      {"type": "nominal", 'title': 'most represented lineage',"field": "most_represented_lineage"},
      {"type": "quantitative", 'title': 'genomes sequenced', "field": "sequences_number"}
    ]

        
    )
)

# Maintenant on affiche notre graphe complet
total_chart = (
    basemap + bubbles # on superpose nos bulles à notre carte coloriée de l'Afrique
).properties(
        title='Most represented SARS-CoV-2 lineages in Africa', # on lui donne un titre
        width=600, # et des dimensions
        height=250
    ).project(
        type='identity', 
        reflectY=True
    )
total_chart.display() # on l'affiche finalement
````

Et voilà le travail ! N'hésitez pas à passer votre souris au dessus du graphe, les informations détaillées par pays s'affichent.

### Calcul de l'arbre

Il existe là aussi de nombreuses façon de faire des arbres. J'ai déjà pu utiliser des outils tels que :

- __[Fasttree](http://www.microbesonline.org/fasttree/)__, outil utilisant la méthode de Maximum Likelihood (ML) approximatif, très rapide bien que soumis à certaines approximations, sur le serveur Galaxy

### Annotations iTOL

Nous allons utiliser __[iTOL](https://itol.embl.de/itol.cgi)__ pour éditer notre arbre ainsi obtenu. iTOL est un outil en ligne qui permet une grande liberté dans son édition et l'ajout d'informations. Vous pouvez trouver son mode d'emploi détaillé __[ici](https://itol.embl.de/help.cgi)__ Ici à notre arbre nous allons modifier trois choses :
- le nom des feuilles qui n'est pas bien lisible, nous allons les transformer en le format *lignée*|*pays*|*date*|*numéro d'accession*
- colorier le label de la feuille en fonction du pays dans lequel sa lignée est la plus représentée
- ajouter une barre affichant le nombre de séquences de la lignée trouvée en Afrique

La fonction suivante permet de créer un fichier d'annotation iTOL sur la base d'un __[template LABELS](https://itol.embl.de/help/labels_template.txt)__. Cela permet de changer le texte attribué aux nœuds et feuilles. Ici on va changer les labels des feuilles qui sont sous forme de numéro d'accession NCBI en un format "lignée|pays|date|numéro d'accession NCBI". Pour ceci, on va s'aider de notre dataset.

````python
# fonction pour créer un fichier d'annotation iTOL pour changer les labels en un format lignée|pays|date|numéro d'accession NCBI
def itols_labels_editor(input_df, outputfile):
    with open(outputfile, 'w') as file:
        file.write("LABELS\n")
        file.write("SEPARATOR TAB\n")
        file.write("\n")
        file.write("DATA\n")

        new_labels = [f"{w}\t{x}|{y}|{z.date()}|{w}" for w, x, y, z in zip(input_df['accession'], 
                                                                input_df['lineage'], 
                                                                input_df['country'], 
                                                                input_df['date'])]
        for element in new_labels:
            file.write(f"{element}\n") # on écrit les données dans un fichier texte
````

La fonction suivante permet de créer un fichier d'annotation iTOL sur la base d'un __[template MULTI_STYLE](https://itol.embl.de/help/label_multistyle.txt)__. Cela permet de fixer des règles de style pour tous les labels de l'arbre. Ici on va mettre en **gras** le premier élément du label (la lignée), mettre en *italique* le second (le pays) et laisser en écriture "normale" les deux derniers éléments (date de prélévement et numéro d'accession NCBI).

````python
# fonction pour créer un fichier d'annotation iTOL pour éditer un peu le style d'écriture des labels
def labels_style_editor(outputfile):
    with open(outputfile, 'w') as file:
        file.write("MULTI_STYLE\n")
        file.write("SEPARATOR COMMA\n")
        file.write("LABEL_SEPARATOR,|\n")

        file.write("\n")
        file.write("DATA\n")
        file.write("1,bold,normal,#000000,1,0,0,default\n")
        file.write("2,italic,normal,#000000,1,0,0,default\n")
        file.write("3,normal,normal,#000000,1,0,0,default\n")
        file.write("4,normal,normal,#000000,1,0,0,default\n")
````

La fonction suivante permet de créer un fichier d'annotation iTOL sur la base d'un __[template DATASET_SIMPLEBAR](https://itol.embl.de/help/dataset_simplebar_template.txt)__. Chaque nœud de l'arbre est associé à une valeur numérique unique qui est affichée sous forme de barre à l'extérieur de l'arbre. Ici on va afficher une barre pour chacune des séquences référence, représentant le nombre de séquences de cette lignée en Afrique.

````python
# fonction pour écrire un fichier d'annotation iTOL permettant de représenter pour chaque lignée, le nombre de 
# séquences collectées en Afrique sous forme d'une barre de hauteur proportionnelle au nombre
def itols_lineage_sequences_number_anno_editor(input_df, most_represented_df, outputfile):
    with open(outputfile, 'w') as file:
        file.write("DATASET_SIMPLEBAR\n")
        file.write("SEPARATOR TAB\n")
        file.write("DATASET_LABEL\tbar label\n")
        file.write("COLOR\t#ff0000\n")
        file.write("SHOW_VALUE\t1\n")
        file.write("LABEL_POSITION\tleft\n")

        file.write("\n")
        file.write("DATA\n")        

        new_labels = [f"{x}\t{
        str(
            most_represented_df.loc[most_represented_df['lineage']==y].genomes_number.to_string(index=False)
            )
        }" for x, y in zip(
            input_df['accession'],
            input_df['lineage'])]
        for element in new_labels:
            file.write(f"{element}\n")   
````

La fonction suivante permet de créer un fichier d'annotation iTOL sur la base d'un __[template DATASET_STYLE](https://itol.embl.de/help/dataset_styles_template.txt)__. Cela permet d'ajouter d'éditer le style du texte des noeuds ou des feuilles. Ici on va colorer le fond du label de nos feuilles en fonction du pays dans lequel la lignée a été la plus retrouvée. 

````python
# fonction pour écrire un fichier d'annotation iTOL permettant de colorer le fond du label de chaque feuille de la couleur
# du pays dans lequel sa lignée a été le plus trouvé (couleurs sélectionnées précédemment)
def itols_lineage_most_represented_anno_editor(input_df, most_represented_df, outputfile):

    category_dict_countries = dict(zip(fixed_domain_countries, fixed_range_countries))
    with open(outputfile, 'w') as file:
        file.write("DATASET_STYLE\n")
        file.write("SEPARATOR TAB\n")
        
        file.write("DATASET_LABEL\tCountry\n")

        file.write("LEGEND_TITLE\tCountry where most present\n")
        legend_shape_list=[]
        legend_colors_list=[]
        legend_labels_list=[]
        legend_shape_scales_list=[]
        new_labels=[]

        for x, y in category_dict_countries.items() :
            legend_shape_list.append("1")
            legend_colors_list.append(y)
            legend_labels_list.append(x)
            legend_shape_scales_list.append("1")  

        file.write(f"LEGEND_SHAPES\t{("\t").join(legend_shape_list)}\n")
        file.write(f"LEGEND_COLORS\t{("\t").join(legend_colors_list)}\n")
        file.write(f"LEGEND_LABELS\t{("\t").join(legend_labels_list)}\n")
        file.write(f"LEGEND_SHAPE_SCALES\t{("\t").join(legend_shape_scales_list)}\n")


        file.write("\n")
        file.write("DATA\n")

        new_labels = [f"{x}\tlabel\tnode\t#000000\t1\tnormal\t{
        str(category_dict_countries[
            most_represented_df.loc[most_represented_df['lineage']==y].most_represented_country.to_string(index=False)
            ])
        }" for x, y, z in zip(
            input_df['accession'],
            input_df['lineage'],
            input_df['country']
        )]

        for element in new_labels:
            file.write(f"{element}\n")
````

Maintenant que l'on a créé toutes nos fonctions, on va les utiliser pour créer nos fichiers d'annotations iTOL. Une fois créés, lorsque l'on est sur la page de notre arbre sur le site iTOL, il suffit de faire un glisser-déposer de nos fichiers d'annotations dans la fenêtre iTOL, cela mettre tout de suite à jour notre arbre.

Tous ces fichiers sont déjà présents dans notre dossier */data/iTOL/* mais on remet le code afin de les créer ci-dessous :

````python
iTOL_repo = data_repo + "iTOL/"

# a chaque fois on donne un nom au fichier de sortie
sequences_label_editor_file = iTOL_repo + 'itol_labels_anno.txt'
# a chaque fois on ne créé le fichier que s'il n'existe pas
if not os.path.exists(sequences_label_editor_file):
    itols_labels_editor(sequences_selection, 
                        sequences_label_editor_file)

sequences_label_style_editor_file = iTOL_repo + 'itol_label_style_anno.txt'
if not os.path.exists(sequences_label_style_editor_file):
    labels_style_editor( 
        sequences_label_style_editor_file
    )

sequences_most_represented_anno_file = iTOL_repo + 'itol_most_represented_anno.txt'
if not os.path.exists(sequences_most_represented_anno_file):
    itols_lineage_most_represented_anno_editor(sequences_selection, 
                                               lineages_data, 
                                               sequences_most_represented_anno_file)

sequences_number_anno_file = iTOL_repo + 'itol_sequences_number_anno.txt'
if not os.path.exists(sequences_number_anno_file):
    itols_lineage_sequences_number_anno_editor(sequences_selection, 
                                               lineages_data, 
                                               sequences_number_anno_file)
````
